```python
N = 10 #prime N potenze
K = 2
esponente = 0
potenze = []

while esponente < N:
    potenza = K**esponente
    potenze.append(potenza)
    print ("la base",K,"con esponete", esponente,"ha potenza", potenza)
    esponente += 1
print (potenze)
```

```python
parole = ["Albergo", "Sedia", "Borgo", "Petalo", "Eremo", "Belvedere", "Semestre", "Esteta", "Sosta", "Orpello", "Abete", "Orologio", "Cesta", "Ermellino"]

for parola in parole:
    conteggio = parola.count("e") + parola.count("E")
    print(parola, "-", conteggio)
```

```python
for parola in parole:
    conteggio = parola.lower().count("e")
    print(parola, "-", conteggio)
```
