```python
#Input: una lista di numeri + un numero K
#Output: la media dei numeri nella lista che sono >= K
#Se non ce ne sono, stampare un messaggio appropriato

numeri = [10, 25, 30, 5, 40, 15]
k = 20

numeri_filtrati = []

for numero in numeri:
    if numero >= k:
        numeri_filtrati.append(numero)
print(numeri_filtrati)
```

```python
import pandas as pd

numeri_filtrati = pd.Series(numeri_filtrati)
numeri_filtrati.mean()
```

```python
if len(numeri_filtrati) == 0:
    print("Nessun numero maggiore o uguale a K")
else:
    numeri_filtrati_series = pd.Series(numeri_filtrati)
    print(f"La media è: {numeri_filtrati_series.mean()}")
```

```python
#Cambiare simbolo € in $

prezzi = ["100 €", "200 €", "500 €", "10 €", "50 €", "70 €"]

prezzi_dollari = []

for prezzo in prezzi:
    new_prezzo = prezzo.replace('€', '$')
    prezzi_dollari.append(new_prezzo)
print(prezzi_dollari)
```

```python
#Confronto guadagni con media precedenti

guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

for i in range(len(guadagni)):
    if i == 0:
        print(f"Mese 1: {guadagni[0]} €")
    else:
        media_prec = sum(guadagni[0:i]) / i
        guadagno_attuale = guadagni[i]
        
        if guadagno_attuale > media_prec:
            confronto = "il guadagno attuale è maggiore"
        else:
            confronto = "il guadagno attuale è minore"
        
        print(f"Mese {i+1}: {guadagno_attuale} € (media prec: {media_prec:.2f} € - {confronto})")
```
