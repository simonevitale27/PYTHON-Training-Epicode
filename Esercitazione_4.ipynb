```python
df = df.set_index(df.columns[0])
df.head()
```

```python
#Quali sono i Pokémon leggendari?

df['Legendary'].unique()
```

```python
filtro = df['Legendary'] == True
leggendari = df[filtro]
leggendari.shape
```

```python
leggendari.head()
```

```python
#quali sono i leggendari di tipo 1 Grass?

filtro = df['Legendary'] == True
filtro_2 = df['Type 1'] == 'Grass'
leggendari_grass = df[filtro & filtro_2]

leggendari_grass.shape
leggendari_grass.head()
```

```python
#E leggendari di tipo 1 Ice o Fire?

filtro_leg = df['Legendary'] == True
filtro_ice = df['Type 1'] == 'Ice'
filtro_fire = df['Type 1'] == 'Fire'

filtro_tot = df[filtro_leg & (filtro_ice | filtro_fire)]
filtro_tot.shape
filtro_tot.head()
```

```python
#Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?

filtro_gen = df['Generation'] == 1
filtro_atk = df['Attack'] > 50
filtro_hp = df['HP'] < 60

filtro_tot = df[filtro_gen & filtro_atk & filtro_hp]
filtro_tot.shape
filtro_tot.head()
```
