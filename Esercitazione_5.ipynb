```python
import os

import dotenv
import pandas as pd
import sqlalchemy

dotenv.load_dotenv(dotenv_path="credenziali.env", override=True)

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(connection_string)

db_engine = sqlalchemy.create_engine(connection_string)
```

```python
query = "SELECT * FROM dimproduct"

df = pd.read_sql(query, db_engine)
df.head()
```

```python
#Media del peso dei prodotti
#Peso minimo
#Peso massimo
#Quanti prodotti pesano più di 100?


media_peso = df['Weight'].mean().round()
peso_min = df['Weight'].min()
peso_max = df['Weight'].max()
mag = (df['Weight'] > 100)

print(f'Media peso prodotti: {media_peso}, peso min: {peso_min}, peso max: {peso_max}, i prodotti più pesanti di 100 sono: {mag.sum()}')
```

```python
#Quanto costano in media i prodotti (colonna DealerPrice)?

media = df['DealerPrice'].mean().round(2)

print(f'Costano in media: {media}')
```

```python
#Quali sono i quartili dei prezzi?

df['DealerPrice'].describe()
```

```python
#Prezzo medio per prodotti di colore blu?

filtro_col = df['Color'] == 'Blue'
prodotti_blu = df[filtro_col]
print(f'Prezzo medio prodotti blu: {prodotti_blu['DealerPrice'].mean().round(2)}')
```

```python
#Prezzo medio per prodotti di colore rosso o nero?

filtro_rosso = df['Color'] == 'Red'
filtro_nero = df['Color'] == 'Black'
prodotti = df[filtro_rosso | filtro_nero]
print(f'Prezzo medio prodotti blu: {prodotti['DealerPrice'].mean().round(2)}')
```

```python
#Prezzo massimo per prodotti di taglia 42 e peso oltre i 10 Kg?

taglia = df['Size'] == 42
peso = df['Weight'] > 10
prodotti = df[taglia & peso]

print(f'Prezzo massimo dei prodotti è: {prodotti['DealerPrice'].max()}')
```

```python
#Nome inglese e StandardCost di tutti i prodotti di taglia 42, peso oltre 10 Kg e colore argento?

colore = df['Color'] == 'Silver'
size = df['Size'] == 42
peso = df['Weight'] > 10

prodotti = df[colore & size & peso]

print(prodotti[['EnglishProductName', 'StandardCost']])
```

```python
#Visualizzare StandardCost e DealerPrice degli ultimi 20 elementi: c'è un pattern? Qualche elemento non lo segue?

ultimi = df.tail(20)

print(ultimi[['StandardCost', 'DealerPrice']])

#gli ultimi 3 prodotti hanno un dealerprice più basso dello standard cost, l'aziende vende quindi in perdita
```

```python
def assegna_quartile(bmi):
    if bmi < 26.29:
        return 'Q1'
    elif bmi < 30.4:
        return 'Q2'
    elif bmi < 34.69:
        return 'Q3'
    else:
        return 'Q4'

df['bmi_quartile'] = df['bmi'].apply(assegna_quartile)

df.groupby('bmi_quartile')['charges'].agg(['min', 'mean', 'max'])
```

```python
#Esercizio 11

# DataFrame dipendenti
employees_df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'department_id': [1, 2, 1, 2, 3]
})

# DataFrame dipartimenti
departments_df = pd.DataFrame({
    'department_id': [1, 2, 3],
    'department_name': ['HR', 'IT', 'Finance'],
    'location': ['New York', 'San Francisco', 'Chicago']
})

print(employees_df)
print(departments_df)
```

```python
#Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni sia sui dipendenti che sui dipartimenti, usando la funzione

merge_df = pd.merge(left=employees_df, right=departments_df, on='department_id')
merge_df
```

```python
#Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente; poi unire i due dataset mediante il metodo .join()

new_employees_df= employees_df.set_index('department_id')
new_employees_df
```

```python
new_departments_df= departments_df.set_index('department_id')
new_departments_df
```

```python
#Esercizio 12

#Dal database AdventureWorksDW importiamo le tabelle dimemployee e dimemployeesalesterritory come DataFrame

import os

import dotenv
import pandas as pd
import sqlalchemy

dotenv.load_dotenv(dotenv_path="credenziali.env", override=True)

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(connection_string)

db_engine = sqlalchemy.create_engine(connection_string)
```

```python
query_1 = "SELECT * FROM dimemployee"
df_employee = pd.read_sql(query_1, db_engine)

query_2 = "SELECT * FROM dimemployeesalesterritory"
df_sales_territory = pd.read_sql(query_2, db_engine)
```

```python
df_employee.shape
```

```python
df_sales_territory.shape
```

```python
df_employee.columns
```

```python
df_sales_territory.columns
```

```python
df_sales_territory.head()
```

```python
df_employee.head()
```

```python
df_merge = pd.merge(left=df_employee, right=df_sales_territory, on='EmployeeKey')
df_merge
```

```python
#Importiamo ora la tabella dimsalesterritory ed effettuiamo un join tra questa e il DataFrame risultante della join precedente, usando le colonne SalesTerritoryKey

query_1 = "SELECT * FROM dimsalesterritory"
df_dimsalesterritory = pd.read_sql(query_1, db_engine)
df_dimsalesterritory.head()
```

```python
df_merge_terr = pd.merge(left=df_merge, right=df_dimsalesterritory, on='SalesTerritoryKey')
df_merge_terr
```

```python
#Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)

df_merge_terr.groupby('SalesTerritoryCountry')['EmployeeKey'].count()
```

```python
df_merge_terr.groupby('SalesTerritoryRegion')['EmployeeKey'].count()
```

```python
#Valutiamo la media del BaseRate per ogni paese: ci sono differenze?

df_merge_terr.groupby('SalesTerritoryCountry')['BaseRate'].mean()
```
